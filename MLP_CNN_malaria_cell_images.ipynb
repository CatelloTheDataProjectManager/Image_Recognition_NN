````python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
import matplotlib.pyplot as plt

# Chemin vers le dataset
dataset_path = "C:/Users/cated/Documents/dev/IA school/Deep Learning/Projet/unzipped_files/cell_images_subset"

# Paramètres
img_height, img_width = 150, 150
batch_size = 32

# Chargement des images
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.3)

train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

test_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Conception du modèle MLP
mlp_model = Sequential([
    Flatten(input_shape=(img_height, img_width, 3)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(train_generator.num_classes, activation='softmax')
])

# Compilation du modèle MLP
mlp_model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Entraînement du modèle MLP
mlp_history = mlp_model.fit(
    train_generator,
    epochs=5,
    validation_data=test_generator
)

# Évaluation du modèle MLP
mlp_loss, mlp_accuracy = mlp_model.evaluate(test_generator)
print(f"MLP Accuracy: {mlp_accuracy}")

# Conception du modèle CNN
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(train_generator.num_classes, activation='softmax')
])

# Compilation du modèle CNN
cnn_model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Entraînement du modèle CNN
cnn_history = cnn_model.fit(
    train_generator,
    epochs=5,
    validation_data=test_generator
)

# Évaluation du modèle CNN
cnn_loss, cnn_accuracy = cnn_model.evaluate(test_generator)
print(f"CNN Accuracy: {cnn_accuracy}")

# Tracé des courbes d'apprentissage
def plot_learning_curves(history, title):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

# Tracé des courbes pour MLP et CNN
plot_learning_curves(mlp_history, 'MLP')
plot_learning_curves(cnn_history, 'CNN')
````

## Description des Techniques Utilisées

### Optimiseur : Adam
Adam est un algorithme qui ajuste automatiquement le taux d'apprentissage de chaque paramètre pendant l'entraînement du modèle pour accélérer la convergence.

### Fonction de perte : Entropie croisée catégorielle (Categorical Cross-Entropy)
C'est une fonction de perte utilisée pour les tâches de classification. Elle mesure la différence entre les prédictions du modèle et les vraies valeurs.

### Couches de convolution et max-pooling (Convolutional Layers and Max-Pooling)
- **Couches de convolution** : Elles appliquent des filtres pour détecter des caractéristiques comme les bords et les textures dans les images.
- **Max-pooling** : Réduit la taille des images tout en conservant les informations importantes en prenant le maximum d'une fenêtre de pixels.

### Couche dense avec activation ReLU (Dense Layer with ReLU Activation)
- **Couche dense** : Tous les neurones sont connectés entre eux.
- **ReLU** : Transforme les valeurs négatives en zéro, ce qui aide le modèle à apprendre des relations non-linéaires.

### Couche de sortie avec activation softmax (Output Layer with Softmax Activation)
- **Softmax** : Convertit les sorties en probabilités pour chaque classe, utiles pour la classification.

### Couche d'aplatissement (Flatten Layer)
Elle transforme les données en un vecteur unidimensionnel pour pouvoir les passer aux couches denses.

## Choisir la Meilleure Époque avec le Sauvegarde de Modèle

Dans ce projet, nous utilisons le callback `ModelCheckpoint` de Keras pour sauvegarder la meilleure version de nos modèles pendant l'entraînement. Cela garantit que le modèle avec la meilleure précision de validation est conservé, ce qui peut être crucial pour des performances optimales.

### Comment ça marche

1. **Configuration du Point de Contrôle** :
   Nous définissons un chemin de sauvegarde où le meilleur modèle sera enregistré. Le callback `ModelCheckpoint` surveille la précision de validation (`val_accuracy`) pendant l'entraînement. Il sauvegarde le modèle uniquement lorsqu'il y a une amélioration.

   ```python
   checkpoint_path = "best_model.h5"

````python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt

# Définition des paramètres
dataset_path = "C:/Users/cated/Documents/dev/IA school/Deep Learning/Projet/unzipped_files/cell_images_subset"
img_height, img_width = 150, 150
batch_size = 32
epochs = 5

# Prétraitement des images
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.3)

train_generator = datagen.flow_from_directory(
    dataset_path, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='training'
)

val_generator = datagen.flow_from_directory(
    dataset_path, target_size=(img_height, img_width), batch_size=batch_size, class_mode='categorical', subset='validation'
)

# Callback pour sauvegarder le meilleur modèle
checkpoint_path = "best_model.h5"
checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)

# Définition du modèle MLP
mlp_model = Sequential([
    Flatten(input_shape=(img_height, img_width, 3)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(train_generator.num_classes, activation='softmax')
])

# Compilation et entraînement du modèle MLP
mlp_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
mlp_history = mlp_model.fit(train_generator, epochs=epochs, validation_data=val_generator, callbacks=[checkpoint])

# Évaluation du modèle MLP
mlp_loss, mlp_accuracy = mlp_model.evaluate(val_generator)
print(f"MLP Accuracy: {mlp_accuracy:.4f}")

# Définition du modèle CNN avec Dropout
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(2, 2),
    Dropout(0.2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Dropout(0.2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(train_generator.num_classes, activation='softmax')
])

# Compilation et entraînement du modèle CNN
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_history = cnn_model.fit(train_generator, epochs=epochs, validation_data=val_generator, callbacks=[checkpoint])

# Évaluation du modèle CNN
cnn_loss, cnn_accuracy = cnn_model.evaluate(val_generator)
print(f"CNN Accuracy: {cnn_accuracy:.4f}")

# Fonction pour tracer les courbes d'apprentissage
def plot_learning_curves(history, title):
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.show()

# Tracé des courbes d'apprentissage
plot_learning_curves(mlp_history, 'MLP')
plot_learning_curves(cnn_history, 'CNN')
````
